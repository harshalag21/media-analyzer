```python
csv_id = [6201,7801]
```

```python
def filename(i):
    return f'allsides_news_{i}_need.csv'
```

```python
df = pd.read_csv('allsides_news_5501_need_appended.csv')
df
```

```python
for id in csv_id:
    df = pd.concat([df, pd.read_csv(filename(id))], ignore_index=True, sort=False)
```

```python
df.to_csv(f'allsides_news_{csv_id[len(csv_id)-1]}_need_appended.csv', index=False)
```

```python
# remove datapoints with no heading
df = df[~df['heading'].isnull()].reset_index(drop=True)
df
```

```python
df = df.drop(columns=['Unnamed: 0'], axis=1)
df
```

```python
import datetime
```

```python
df = df[df['date']<datetime.date(2024,7,1)]
```
